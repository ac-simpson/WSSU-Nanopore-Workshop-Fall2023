## Passing data to phyloseq

Now that we've done some taxonomic classification of our reads using different ways of filtering our sequencing data, let's get the data into a format that is more easily explored. We'll convert our data into biom files.

What is a biom file?
http://biom-format.org/ Known as a biological observation matrix, it's a very handy format for storing information about microbial taxa detected in metagenomic data and how many reads are assigned to each taxon. It was designed so that you can use it with a very wide variety of different applications and pipelines.

Now, let's use a program called kraken-biom (https://github.com/smdabdoub/kraken-biom) to convert our reports into one big biom file. We can put all our results into the file and each file will be treated as a separate sample.

```bash
cd bracken_results
```

```bash
kraken-biom no_filter.txt length_filter.txt quality_filter.txt quality_length_filter.txt myfilter.txt \
    --fmt json -o ../Zymo_16S_table.biom
```
